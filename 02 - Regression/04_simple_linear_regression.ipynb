# Parte 2 - Sección 4: Regresión lineal simple

```python
import pandas as pd

df = pd.read_csv('Salary_Data.csv')
df.head()
```

**Descripción del problema y dataset**
El dataset contiene información sobre el salario de los empleados en función de sus años de experiencia.
El objetivo es predecir el salario de un empleado en función de sus años de experiencia con ayuda de un modelo de regresión lineal simple.

```python
import numpy as np

# Separar variables dependientes e idenpendientes
X : np.ndarray = df.iloc[:, :-1].values
y : np.ndarray = df.iloc[:, -1].values
```

```python
from sklearn.model_selection import train_test_split

# Separar los datos para entrenar el modelo y para despues testearlo
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 1/3, random_state= 0)
```

```python
from sklearn.linear_model import LinearRegression

regressor = LinearRegression() # Definir modelo No ENTRENADO
regressor.fit(X = X_train, y = y_train) # Entrenar modelo
print('¡Modelo de regresión simple entrenado!')
```

```python
y_pred = regressor.predict(X_test) # Predecir modelo con los valores de prueba

pd.DataFrame({'y_test': y_test, 'y_pred':y_pred}) # Visualizar valores
```

```python
import matplotlib.pyplot as plt

plt.scatter(x = X_test, y = y_test, color = 'red')
plt.plot(X_test, y_pred, color = 'blue')
plt.title('Sueldo en relación con los años de experencia')
plt.xlabel('Años de experencia')
plt.ylabel('Sueldo')
plt.show()
```

```python
# Valores con los que se entrena el modelo
plt.scatter(x = X_train, y = y_train, color = 'red')
plt.plot(X_train, regressor.predict(X_train), color = 'blue') # == a la linea anterior, porque el modelo se entreno con estos valores
plt.title('Sueldo en relación con los años de experencia')
plt.xlabel('Años de experencia')
plt.ylabel('Sueldo')
plt.show()
```
