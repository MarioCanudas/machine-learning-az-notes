Puede parecer en mismo código que la regla de apriori, y lo es, solo que las reglas se ordenan por el soporte y no por el lift.

```python
import pandas as pd

df = pd.read_csv('Market_Basket_Optimisation.csv', header = None)

# Extraer transacciones por usuario
transactions = []
for i in range(0, len(df)):
    transactions.append([str(df.values[i, j]) for j in range(0, 20) if not pd.isna(df.values[i, j])])
```

```python
from apyori import apriori

# Entrenar modelo
rules = apriori(transactions, min_support = 0.003, min_confidence = 0.2, min_lift = 3, min_lenght = 2, max_lenght = 2)
```

```python
def view_rules(_rules) -> pd.DataFrame:
    if not isinstance(_rules, list):
        _rules = list(_rules)
    
    structured_rules = []
    for r in _rules:
        structured_rules.append(
            {
                'Product 1': tuple(r[2][0][0])[0],
                'Product 2': tuple(r[2][0][1])[0],
                'Support': r[1],
            }
        )

    return pd.DataFrame(structured_rules).drop_duplicates()

df_rules = view_rules(rules)
```

```python
df_rules.nlargest(n = 10, columns = 'Support')
```
